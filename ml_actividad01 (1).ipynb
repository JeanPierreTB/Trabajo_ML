5. Ordena el DataFrame por 'fecha_entrega' en orden ascendente. (0.5pts)

```python
import pandas as pd
from datetime import datetime



def convertir_fecha(fecha_texto):
    meses = {
        'Enero': '01', 'Febrero': '02', 'Marzo': '03', 'Abril': '04',
        'Mayo': '05', 'Junio': '06', 'Julio': '07', 'Agosto': '08',
        'Septiembre': '09', 'Octubre': '10', 'Noviembre': '11', 'Diciembre': '12'
    }
    
    
    partes = fecha_texto.split(' ')
    dia = partes[0]
    mes = meses[partes[2][:-1]]  
    año = partes[3]
    
   
    fecha_formateada = f"{dia}/{mes}/{año}"
    return fecha_formateada

dataset['fecha_entrega'] = dataset['fecha_entrega'].apply(convertir_fecha)

dataset['fecha_entrega'] = pd.to_datetime(dataset['fecha_entrega'], format='%d/%m/%Y')

dataset=dataset.sort_values(by='fecha_entrega')

dataset=dataset.reset_index(drop=True)

dataset






```

6. Agrega una nueva columna 'precio_por_area' que calcule el precio por metro cuadrado (precio / area). (1 pts)

```python
# tu código aquí


dataset['area']=dataset['area'].str.replace('m2','',regex=False)

dataset['area']=dataset['area'].astype('float')



precio2= dataset['precio'].str.replace('S/ ', '', regex=False)  
precio2 = precio2.str.replace(',', '', regex=False)  
precio2 = precio2.str.replace('$ ', '', regex=False)   

precio2 = precio2.astype(float)

print("Precio:",precio2)


dataset['precio_por_area']=precio2/dataset['area']


dataset

```

10. A partir del dataset y las columnas, utilizar un modelo de Regresión Linear de ScikitLearn para encontrar el modelo que pueda explicar la correlación entre `precio` y `area`. (4 pts)

Considerando la siguiente ecuación
$Y = X*w + b$

$Y$: `valor del precio`
$X$: `tamaño del departamento en m2` 

$w$ y $b$: coeficientes 


```python
#tu código aquí

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

X=dataset['area']
Y=dataset['precio']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)

X_train_reshaped=X_train.values.reshape(-1,1)
Y_train_reshaped=Y_train.values.reshape(-1,1)
X_test_reshaped=X_test.values.reshape(-1,1)

model = LinearRegression()
model.fit(X_train_reshaped, Y_train_reshaped)


Y_pred = model.predict(X_test_reshaped)

```

```python
#imprime tus parámetros w y b aquí

print("Coeficiente (w):", model.coef_[0])
print("Intersección (b):", model.intercept_)
```

11.  ¿Cómo varía los coeficientes `w` y `b` dependiendo del distrito? Seleccione 5 casos y analice `w` y `b`. (2 pts)

```python
# tu código aquí
# Lista de distritos a analizar
distritos_a_analizar = ['Barranco', 'Ate', 'Comas', 'Lince', 'San Borja']

coeficientes_por_distrito = {}

for distrito in distritos_a_analizar:
    datos_distrito = dataset[dataset['distrito'] == distrito]
    
    X = datos_distrito['area']
    Y = datos_distrito['precio']
    
    X_reshaped = X.values.reshape(-1, 1)
    Y_reshaped = Y.values.reshape(-1, 1)
    
    model = LinearRegression()
    model.fit(X_reshaped, Y_reshaped)
    
    coeficientes_por_distrito[distrito] = {
        'coeficiente (w)': model.coef_[0][0],
        'intersección (b)': model.intercept_[0]
    }

coeficientes_df = pd.DataFrame(coeficientes_por_distrito).T

print(coeficientes_df)
```
